# Part 1
Fixing the columns for نوع التمويل, but since it's heavely correlated with the المنتج column I'll start with that first.

إجتماعي and تمويل الافراد are the same

مشروع and تمويل الاعمال are the same 

نقل and تمويل نقل are the same

حر contains two values:

-> قرض العمل الحر نقدي belongs to تمويل الاعمال/مشروع
                       
-> قرض النقل التشاركي belongs to تمويل نقل/نقل

```python
data["نوع_التمويل"] = data["نوع_التمويل"].replace(['إجتماعي','تمويل الافراد'],'تمويل الافراد')
data["نوع_التمويل"] = data["نوع_التمويل"].replace(['مشروع','تمويل الاعمال'],'تمويل الاعمال')
data["نوع_التمويل"] = data["نوع_التمويل"].replace(['نقل','تمويل نقل'],'تمويل نقل')

for index, row in data.iterrows():
    if row["المنتج"] == "قرض العمل الحر نقدي":
        data["نوع_التمويل"][index] = "تمويل الاعمال"
    elif row["المنتج"] == "قرض النقل التشاركي":
        data["نوع_التمويل"][index] = "تمويل نقل"
```

# Part 2
Fixing the قيمة_الدخل column.

Every value here has two versions, one with space and another without. 

So we'll first unify them by choosing the one without a space. (Thank you abdulrahman!)

```python
data.loc[data["قيمة_الدخل"].isin(['< 5000', '<5000']), 'قيمة_الدخل'] = '< 5000'
data.loc[data["قيمة_الدخل"].isin(['>= 7500', '>=7500']), 'قيمة_الدخل'] = '>= 7500'
data.loc[data["قيمة_الدخل"].isin(['>= 10000', '>=10000']), 'قيمة_الدخل'] = '>= 10000'
data.loc[data["قيمة_الدخل"].isin(['>= 5000', '>=5000']), 'قيمة_الدخل'] = '>= 5000'
data["قيمة_الدخل"].fillna('غير معروف', inplace = True)
```

the catagories here is a bit confusing so I'm going to create a new column named **نوع_الدخل** which will catagorize the values into a more readable language

<5000 will be **منخفض جدا**

\>=5000 will be **منخفض**

\>=7500 will be **متوسط**

\>=10000 will be **مرتفع**

```python
def income_type(row):
    if row["قيمة_الدخل"] == "< 5000":
        return "منخفض جدا"
    elif row["قيمة_الدخل"] == ">= 5000":
        return "منخفض"
    elif row["قيمة_الدخل"] == ">= 7500":
        return "متوسط"
    elif row["قيمة_الدخل"] == ">= 10000":
        return "مرتفع"
    return "غير معروف"
    
#data["نوع_الدخل"] = 
data.insert(len(data.columns)-1, "نوع_الدخل", data.apply(lambda row: income_type(row), axis=1)) # using insert to insert the new column in the second place from the end.
                                                                                                # so the year stays at the end
```
